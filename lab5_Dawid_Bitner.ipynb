```python
import pandas as pd

# Zadanie 1
teachers = pd.read_csv("C:\\Users\\Dawid\\Desktop\\semestr6\\jezyki skryptowe w analizie danych\\nauczyciele.csv", sep=";", decimal=",", index_col="Nazwa")
teachers
```

### zad 2

```python
import numpy as np

teachers_ready = teachers.drop(["Unnamed: 102", "Kod"], axis=1).drop(["POLSKA"], axis=0)
cols = np.array([col.split(';')[:-1] for col in teachers_ready.columns]).transpose()
t_multiindex = pd.MultiIndex.from_arrays(cols, names=["Kto", "Plec", "Gdzie", "Rok"])
teachers_ready.columns = t_multiindex
teachers_ready
```

### zad 3

```python
import matplotlib.pyplot as plt

print("a)")
data = teachers_ready.xs(['nauczyciele akademiccy', 'ogółem'], level=('Kto', 'Plec'), axis=1)["uniwersytety"] + teachers_ready.xs(['nauczyciele akademiccy', 'ogółem'], level=('Kto', 'Plec'), axis=1)["wyższe szkoły techniczne"]

plt.plot(range(len(data)), data,'o')
plt.xticks(range(len(data.index)), data.index, rotation=90)
plt.title('Akademiccy nauczyciele zatrudnieni w latach 2014-2018')
plt.ylabel('Ilość osób', fontsize=15)
plt.legend([2014, 2015, 2016, 2017, 2018], bbox_to_anchor=(1.4, 1))

plt.show()
plt.close()

print("b)")
x_ticks = list(dict.fromkeys(teachers_ready.columns.get_level_values(0)))
data = list()
for el in x_ticks:
    data.append((teachers_ready.xs([el, 'ogółem'], level=('Kto', 'Plec'), axis=1)["uniwersytety"] + teachers_ready.xs([el, 'ogółem'], level=('Kto', 'Plec'), axis=1)["wyższe szkoły techniczne"]).loc["ŚLĄSKIE"])
    
for i in range(len(x_ticks)):
    x_ticks[i] = " ".join(x_ticks[i])
    
plt.plot(range(len(data)), data, 'o')
plt.xticks(range(len(x_ticks)), x_ticks, rotation=90)
plt.title('Wszyscy nauczyciele zatrudnieni w latach 2014-2018 w woj. śląskim')
plt.ylabel('Ilość osób', fontsize=15)
plt.legend([2014, 2015, 2016, 2017, 2018], bbox_to_anchor=(1.4, 1))

plt.show()
plt.close()

print("c)")

men = (teachers_ready.xs(['nauczyciele akademiccy'], level=('Kto'), axis=1)["ogółem"] - teachers_ready.xs(['nauczyciele akademiccy'], level=('Kto'), axis=1)["kobiety"]).loc["ŚLĄSKIE"]
women = (teachers_ready.xs(['nauczyciele akademiccy'], level=('Kto'), axis=1)["kobiety"]).loc["ŚLĄSKIE"]

x_ticks = list()
for el in men.index:
    x_ticks.append(" ".join(el))

plt.plot(range(len(women)), women, 'o')
plt.plot(range(len(men)), men, 'o')
plt.xticks(range(len(x_ticks)), x_ticks, rotation=90)
plt.title('Wszyscy nauczyciele zatrudnieni w latach 2014-2018 w woj. śląskim')
plt.ylabel('Ilość osób', fontsize=15)
plt.legend(["Kobiety", "Mezczyzni"], bbox_to_anchor=(1.4, 1))

plt.show()
plt.close()
```

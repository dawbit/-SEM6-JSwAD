#### 1

```r
library(dplyr)
df = read.csv('imports-85.data', header=FALSE, stringsAsFactors=FALSE)
```

#### 2

```r
df = df %>%
    rename(
        symboling = 1,
        normalized_losses = 2,
        make = 3,
        fuel_type = 4,
        aspiration = 5,
        num_of_doors = 6,
        body_style = 7,
        drive_wheels = 8,
        engine_location = 9,
        wheel_base = 10,
        length = 11,
        width = 12,
        height = 13,
        curb_weight = 14,
        engine_type = 15,
        num_of_cylinders = 16,
        engine_size = 17,
        fuel_system = 18,
        bore = 19,
        stroke = 20,
        compression_ratio = 21,
        horsepower = 22,
        peak_rpm = 23,
        city_mpg = 24,
        highway_mpg = 25,
        price = 26,
    )

for(i in names(df)){
    df = df %>%
        filter(!!as.symbol(i) != "?")
}

df
```

#### 3

```r
df = arrange(df, desc(as.numeric(price)))
df[1,]
```

```r
df = transform(df, price = as.numeric(price))

print(mean(df$price))
df %>%
    filter(price>=mean(df$price))
```

```r
df %>%
    filter(price>=mean(df$price) & make=="audi")
```
